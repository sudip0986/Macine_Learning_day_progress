```python
import numpy as np
class GDRegressor:
    def __init__(self, larning_rate, epochs):
        self.coef_= np.random.randn()
        self.intercept_= np.random.randn()
        self.lr=larning_rate
        self.epoch=epochs

    def fit(self, X, y):

        for i in range(self.epoch):

            # Calculate gradients (partial derivatives)
            # slope between L and B = ∂L/∂b = -2 * Σ(y - mx - b)
            slope_b = -2 * np.sum(y - self.coef_ * X.ravel() - self.intercept_)

            # ∂L/∂m = -2 * Σ(y - mx - b) * x
            slope_m = -2 * np.sum((y - self.coef_ * X.ravel() - self.intercept_) * X.ravel())

            # Update parameters
            self.intercept_ = self.intercept_ - self.lr*slope_b
            self.coef_ = self.coef_ -  self.lr*slope_m
            print(f'{i} numer epoch & coef_:{self.coef_} & intersept_:{self.intercept_}')
            
    def predict(self, X):
        return self.coef_*X.ravel() + self.intercept_

    
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20,random_state=42)
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Training Model 
model = GDRegressor(larning_rate=0.01, epochs=100)
model.fit(X_train,y_train)
```

```python
y_pred_myclass = model.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_myclass)
```
